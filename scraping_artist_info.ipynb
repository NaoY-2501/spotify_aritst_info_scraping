```python
import json
import re

import requests
from bs4 import BeautifulSoup as bs
```

```python
# artist_id = '11Z78hOagXfA82mOlT6baL'  # 茶太
# artist_id = '13tttiaoECFPQKEigVnYU6'  # みゆはん
artist_id = '2dEKl1tM3JMwggawEuVISO'  # sora tob sakana
# artist_id = '2rkgBTtk0Ec32yjIYsFP87'  # Maison book girl
# artist_id = '4L5K9BR3qaZQw8QAoMBUta'  # にゃーろんず
```

```python
url = 'https://open.spotify.com/artist/{artist_id}/about'.format(artist_id=artist_id)
```

```python
res = requests.get(url)
```

```python
res.content
```

```python
res.headers
```

```python
soup = bs(res.content)
```

```python
scripts = soup.find_all('script')
scripts
```

```python
script_json = scripts[3]
script_json
```

```python
artist_json = json.loads(script_json.text.strip())
```

```python
artist_json
```

```python
artist_json.keys()
```

```python
desc = artist_json['description']
desc
```

```python
MONTHLY = re.compile(r'Monthly Listeners:[\s][\d]+')
```

```python
monthly = re.search(MONTHLY, desc)[0]
monthly
```

```python
monthly_listeners = monthly.split(':')[1].strip()
monthly_listeners
```

```python
WHERE = re.compile(r'Where People Listen:[\s\S]+')
```

```python
cities = re.search(WHERE, desc)[0]
cities
```

```python
cities.replace('Where People Listen: ', '').split(',')
```
